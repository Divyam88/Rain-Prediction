# Checking the NAN values in columns

```python
df.isnull().sum()
```

# Dealing with Null values

```python
columns = list(df.columns.values)
for col in columns:
    most_frequent_values = df[col].value_counts().idxmax()
    df[col].replace(np.nan, most_frequent_values, inplace=True)
```

```python
df.isnull().sum()
```

# Correlation between the features

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(14, 14))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
df.corr()
```

```python
df.columns
```

```python
# Dropping the Unnecessary columns

df.drop(['Date', 'Location'], axis =1, inplace=True)
df.head()
```

# Feature Encoding

Since, machine learning models accepts numerical values. We need to convert the columns with type "object" to a numerical type.

```python
# Checking for categorical columns

categorical_cols = df.columns[df.dtypes == object]
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for feat in categorical_cols:
    df[feat] = le.fit_transform(df[feat].astype(str))
```

```python
df[categorical_cols].head(10)
```

# Splitting the dataset into training and testing set


```python
# Splittng the dataset into independent and dependent variables
X = df.drop(['RainTomorrow'], axis=1)
Y = df.RainTomorrow
```

# Feature Scaling
To Bring the features in one range of values.

```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
X = s.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

# Training Model
After analysing the dataset, we can say that it is a classification problem, since, we have to predict True or False.

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=100,random_state=0)
classifier.fit(X_train,Y_train)
```

```python
classifier.score(X_train,Y_train)
```

```python
y_pred = le.inverse_transform(np.array(classifier.predict(X_test),dtype=int))
Y_test = le.inverse_transform(np.array(Y_test,dtype=int))
```

```python

print(y_pred)
```

```python

print(Y_test)
```

```python
y_pred = y_pred.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
```

```python
df = np.concatenate((Y_test,y_pred),axis=1)
dataframe = pd.DataFrame(df,columns=['Rain on Tommorrow','Predition of Rain'])
```

```python
print(dataframe)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
